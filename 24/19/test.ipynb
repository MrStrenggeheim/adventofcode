```python
def check(towel):
    global I 
    I += 1
    if I > 25:
        return False
    if len(towel) == 0:
        return True
    for pattern in patterns:
        if towel[:len(pattern)] == pattern:
            if check(towel[len(pattern):]):
                return True
    return False
```

```python
class Node:
    def __init__(self, val = 0):
        self.val = val
        self.next = dict()
            
    def insert(self, word):
        c = word[0]
        if c in self.next.keys():
            if len(word) == 1:
                self.next[c].val = 1
            else:
                self.next[c].insert(word[1:])
        else:
            if len(word) == 1:
                self.next[c] = Node(val=1)
            else:
                self.next[c] = Node()
                self.next[c].insert(word[1:])
    
    def search(self, word):
        if len(word) == 0:
            return self.val == 1
        c = word[0]
        if c in self.next.keys():
            return self.next[c].search(word[1:])
        return False
    
    def get_all_patterns(self, to_scan, scanned=""):
        """get all patterns that match the beignning of word"""
        ret = []
        if self.val == 1:
            ret.append(scanned)
        if to_scan == "":
            return ret
        c = to_scan[0]
        if c not in self.next.keys():
            return ret
        return ret + self.next[c].get_all_patterns(to_scan[1:], scanned=scanned+c)
        
```
