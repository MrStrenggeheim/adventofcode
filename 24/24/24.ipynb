```python
with open("input.txt") as file:
    start_values, gates = file.read().split("\n\n")
```

```python
z_wires = []
dep_dict = dict()
for gate in gates.split("\n"):
    i, o = gate.split(" -> ")
    i1, op, i2 = i.split(" ")
    if o[0] == 'z':
        z_wires.append(o)
    dep_dict[o] = (op, i1, i2)
for sv in start_values.split("\n"):
    wire, value = sv.split(": ")
    dep_dict[wire] = (int(value), )
z_wires.sort()
```

```python
op_map = {
    'AND' : lambda a, b : a & b,
    'OR'  : lambda a, b : a | b,
    'XOR' : lambda a, b : a ^ b
}
```

```python
dep_dict, z_wires
```

```python
class Gate:
    def __init__(self, op, i1, i2):
        self.op = op_map[op]
        i1 = dep_dict[i1]
        i2 = dep_dict[i2]
        self.i1 = Gate(*i1) if len(i1) == 3 else Input(i1[0])
        self.i2 = Gate(*i2) if len(i2) == 3 else Input(i2[0])
    def __call__(self):
        return self.op(self.i1(), self.i2())
class Input:
    def __init__(self, value):
        self.value = value
    def __call__(self):
        return self.value
```

```python
z_gates = []
for z in z_wires:
    z_gates.append(Gate(*dep_dict[z]))
```

```python
class Gate():
    def __init__(self, string):
        i, o = string.split(" -> ")
        i1, op, i2 = i.split(" ")
        self.i = (i1, i2)
        self.o = o
        self.op = op
    def __repr__(self):
        return f"G({self.i}, {self.op}, {self.o})"
    def __lt__(self, other):
        print(self)
        return self.o in other.i
```

```python

gates = [Gate(g) for g in gates.split("\n")]
```

```python
start_values, gates[1]
```

```python
def gate_sort(a, b):
    return -1 if b.o in a.i else 1
```
