# Advent of Code 2023 - Day 6: **Wait For It**

```python
import re
import numpy as np
import matplotlib.pyplot as plt
```

```python
input = "input.txt"
with (open(input, "r")) as f:
    data = f.read().splitlines()
    
data = [[int(x) for x in re.split(r" +", line)[1:]] for line in data]
```

```python
def get_num_wins(time, dist):
    r = np.arange(time-1)+1
    r = np.longlong(r)  
    d = r * (time - r)
    num = np.sum(d > dist)
    return num, d
```

```python
# Task 1
num_wins = 1

for i in range(len(data[0])):
    num, _ = get_num_wins(data[0][i], data[1][i])
    num_wins = num_wins * num

print(num_wins)
```

```python
# Task 2
data2 = [int(''.join(map(str, x))) for x in data]

num, d = get_num_wins(data2[0], data2[1])
```

```python
# Visualisation
plt.plot(d[::10])
plt.plot([0, len(d)/10], [data2[1], data2[1]])
plt.show()
```
